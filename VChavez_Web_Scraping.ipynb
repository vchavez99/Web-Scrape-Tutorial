```
"""
Created by Valeria Chavez
March 17th, 2021
Web-Scraping Tutorial - Sript 1
This script will web-scrape Wes Anderson's Wikipedia page and extract
the text from the table of contents. Wes Anderson is my favorite director.
You might know him from "The Grand Budapest Hotel", "Isle of Dogs", or
"Moonrise Kingdom". I chose to do his Wikipedia page because I recently
got a coffe table book about his movies which is probably the best gift
I've ever gotten. 
"""
#Import libraries
import requests
import bs4

#Here we make a request for the page's content
res = requests.get('https://en.wikipedia.org/wiki/Wes_Anderson')

#Here we make our soup object and select the table of contents text
wes_soup = bs4.BeautifulSoup(res.text,"lxml")
wes_soup.select(".toc")

#Now, let's use a "for loop" to iterate through all the items
#and print only text
for item in wes_soup.select('.toc'):
  print(item.text)
```
